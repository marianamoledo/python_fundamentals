# Estruturas Condicionais Bsicas

## Instruo IF

A instruo `if`  uma das estruturas condicionais mais fundamentais na programao. Ela  usada para tomar decises com base em uma condio booleana (verdadeira ou falsa). Como vai funcionar:

* Se a condio for verdadeira (`true`), o bloco de cdigo que est _indentado_ abaixo da instruo `if` ser executado.
* Caso a condio seja falsa (`false`), o bloco de cdigo dentro do `if` ser ignorado e o programa continuar a execuo normalmente.


Ou seja, o if permite que um bloco de cdigo seja executado se uma condio for verdadeira.

```python
x=30

if x>15:
    print("x  maior que 15")
```

mas e quando queremos verificar vrias condies em sequencia?   
da podemos utilizar o _elif_ (abreviao de "else if")!

## Instruo ELIF

Se a condio do if for falsa, o python verifica a prxima condio do elif e assim por diante... sacou?
Vamos pensar num exemplo ento?
Vamos elaborar um cdigo onde vamos classificar uma pessoa de acordo sua faixa etria:
* Criana: At 12 anos.
* Adolescente: Entre 12 e 17 anos.
* Adulto: A partir de 18 at 60 anos.
* Idoso: A partir de 60 anos.

```python
idade = 22

if idade <12:
    print("Criana")
elif idade <18:
    print ("Adolescente")
elif idade <60:
    print("Adulto")
elif idade >=60:
    print("Idoso")
```

e se no atender a nenhuma das condies anteriores?

## A instruo ELSE

A instruo else serve para pegar todas as situaes que no foram atendidas pelas condies anteriores (os comandos if e elif).  

Ou seja, **ela  como uma "ltima chance" de executar algo, caso nenhuma das condies anteriores tenha sido verdadeira.**

Ela no precisa de uma condio para ser verificada; o bloco de cdigo que est dentro do else vai rodar automaticamente se todas as condies anteriores derem errado.  

Podemos reescrever o exemplo anterior da seguinte forma utilizando o _else_:

```python
idade = 70

if idade < 12:
    print("Criana")
elif idade < 18:
    print("Adolescente")
elif idade < 30:
    print("Jovem")
elif idade < 60:
    print("Adulto")
else:
    print("Idoso")
```

Vamos pensar em mais um exemplo s pra fechar com chave de ouro?  

Vamos criar um cdigo em python utilizando _if_, _elif_ e _else_ para classificar o desempenho de um aluno com base na sua nota e considerando as faixas abaixo:  

* Nota abaixo de 5: Reprovado.

* Nota entre 5 e 6.9: Recuperao.

* Nota entre 7 e 8.9: Bom.

* Nota entre 9 e 10: Excelente.

* Nota fora do intervalo (menor que 0 ou maior que 10): Nota invlida.

```python
nota = float(input("Digite a nota do aluno (0 a 10): "))

if nota < 5: 
    print ("Reprovado.")
elif nota < 6.9:
    print ("Recuperao")
elif nota < 8.9:
    print ("Bom")
elif nota <= 10:
    print ("Excelente")
else:
    print("Nota Invlida") 
```

# Loops em Python

Os loops nos permitem repetir uma ao vrias vezes sem precisar escrever o mesmo cdigo repetidamente. Existem duas maneiras de criar loops em Python:

* _for_  Utilizamos quando sabemos quantas vezes queremos repetir uma ao.
* _while_  Executa repetidamente um bloco de cdigo enquanto uma condio for verdadeira. Diferente do for, ele no tem um nmero fixo de repeties, mas continua at que a condio seja falsa. 

## _for_ loop:

Como dito anteriormente, o for  ideal quando sabemos quantas vezes queremos repetir uma ao. Ele percorre elementos de uma sequncia, como listas, strings ou intervalos numricos.

#### Exemplo
Imprima os nmeros de 1 a 5 usando um loop _for_.

```python
for i in range(1,6):
    print (f"iterao {i}")
```

Se a gente fosse traduzir para a linguagem natural, teramos: _"Para cada nmero de 1 a 5, faa o seguinte: imprima esse nmero."_

Vamos fazer o mesmo desafio de descobrir o nmero secreto s que usando o for loop?

Crie um programa que:

## _while_ loop:

O while no tem um nmero fixo de repeties. Ele continua executando enquanto uma condio for verdadeira. Seria algo tipo esse diagrama de fluxo abaixo:  



#### Exemplo
Imprima os nmeros de 1 a 5 usando um loop _while_.

```python
contador = 1

while contador <= 5:
    print (f"Iterao {contador}")
    contador = contador + 1
```

Sacou? Portanto, o loop while  til quando *no sabemos exatamente quantas vezes o cdigo deve rodar* e queremos que ele continue at que uma condio especfica seja atendida.

Vamos fazer o mesmo desafio de descobrir o nmero secreto s que usando o _while_ loop?

Crie um programa que:
- Escolhe um nmero secreto.
- Pede por um chute do usurio.
- Indica se o usurio acertou ou no.
- Se no acertou, d uma dica, dizendo
    - se o nmero  mais baixo ou mais alto
- Repete isso at 3 vezes!

```python
import random

numero_secreto = random.randint(1, 10) 
tentativas = 0
max_tentativas = 3

print("Tente adivinhar o nmero secreto entre 1 e 10!")

while tentativas < max_tentativas:
    chute = int(input("Digite seu chute: "))
    tentativas += 1

    if chute == numero_secreto:
        print("Parabns! Voc acertou!")
        break
    elif chute < numero_secreto:
        print("O nmero secreto  maior!")
    else:
        print("O nmero secreto  menor!")

    if tentativas == max_tentativas:
        print(f"Fim de jogo! O nmero secreto era {numero_secreto}.")
```

Resumindo da diferena entre _while_ e _for_:
| Estrutura | Quando usar? |
|-----------|-------------|
| `while`   | Quando no se sabe exatamente quantas vezes o loop deve rodar (depende de uma condio). |
| `for`     | Quando se sabe o nmero de iteraes ou ao percorrer elementos de uma sequncia. |
