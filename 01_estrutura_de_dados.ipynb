# Operadores de ComparaÃ§Ã£o ðŸŸ°

Agora vamos ver os operadores comparaÃ§Ã£o em python. Eles sÃ£o usados para comparar dois valores e retornam um **valor booleano** (`True` ou `False`).

| Operador | DescriÃ§Ã£o                  | Exemplo       | Resultado |
|----------|----------------------------|--------------|-----------|
| `==`     | Igual a                     | `5 == 5`     | `True`    |
| `!=`     | Diferente de                 | `5 != 3`     | `True`    |
| `>`      | Maior que                    | `10 > 7`     | `True`    |
| `<`      | Menor que                    | `3 < 8`      | `True`    |
| `>=`     | Maior ou igual a             | `6 >= 6`     | `True`    |
| `<=`     | Menor ou igual a             | `4 <= 2`     | `False`   |

Vamos ver um exemplo? 

```python
a = 10
b = 20

print(a == b)  
print(a != b)  
print(a > b)   
print(a < b)   
print(a >= 10) 
print(b <= 15) 
```
