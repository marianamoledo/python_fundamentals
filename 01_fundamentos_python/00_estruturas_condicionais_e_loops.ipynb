Na vida, tomamos decisões o tempo todo, certo?



Na programação é a mesma coisa! Por exemplo, imagine um semáforo controlando o fluxo de carros: ele decide quando o sinal deve estar verde, amarelo ou vermelho, dependendo do contexto. Esse processo de decisão é o que chamamos de *estrutura condicional*.  

No Python, usamos essas estruturas condicionais para permitir que o programa execute diferentes blocos de código com base em condições específicas. É como ensinar o computador a responder perguntas simples de "sim" ou "não", como:

* "O número é maior que 10?"
* "O usuário está logado?"
* "O saldo é suficiente para a compra?"

A principal ferramenta para isso é o comando _if_, acompanhado de _else_ e _elif_, que permitem construir decisões simples e complexas! 😀

# Estruturas Condicionais Básicas

## Instrução IF

A instrução `if` é uma das estruturas condicionais mais fundamentais na programação. Ela é usada para tomar decisões com base em uma condição booleana (verdadeira ou falsa). Como vai funcionar:

* Se a condição for verdadeira (`true`), o bloco de código que está _indentado_ abaixo da instrução `if` será executado.
* Caso a condição seja falsa (`false`), o bloco de código dentro do `if` será ignorado e o programa continuará a execução normalmente.


Ou seja, o if permite que um bloco de código seja executado se uma condição for verdadeira.

```python
x=30

if x>15:
    print("x é maior que 15")
```

mas e quando queremos verificar vááárias condições em sequencia? 🤔  
daí podemos utilizar o _elif_ (abreviação de "else if")!

## Instrução ELIF

Se a condição do if for falsa, o python verifica a próxima condição do elif e assim por diante... sacou?
Vamos pensar num exemplo então?
Vamos elaborar um código onde vamos classificar uma pessoa de acordo sua faixa etária:
* Criança: Até 12 anos.
* Adolescente: Entre 12 e 17 anos.
* Adulto: A partir de 18 até 60 anos.
* Idoso: A partir de 60 anos.

```python
idade = 22

if idade <12:
    print("Criança")
elif idade <18:
    print ("Adolescente")
elif idade <60:
    print("Adulto")
elif idade >=60:
    print("Idoso")
```

e se não atender a nenhuma das condições anteriores?

## A instrução ELSE

A instrução else serve para pegar todas as situações que não foram atendidas pelas condições anteriores (os comandos if e elif).  

Ou seja, **ela é como uma "última chance" de executar algo, caso nenhuma das condições anteriores tenha sido verdadeira.**

Ela não precisa de uma condição para ser verificada; o bloco de código que está dentro do else vai rodar automaticamente se todas as condições anteriores derem errado.  

Podemos reescrever o exemplo anterior da seguinte forma utilizando o _else_:

```python
idade = 70

if idade < 12:
    print("Criança")
elif idade < 18:
    print("Adolescente")
elif idade < 30:
    print("Jovem")
elif idade < 60:
    print("Adulto")
else:
    print("Idoso")
```

Vamos pensar em mais um exemplo só pra fechar com chave de ouro?  

Vamos criar um código em python utilizando _if_, _elif_ e _else_ para classificar o desempenho de um aluno com base na sua nota e considerando as faixas abaixo:  

* Nota abaixo de 5: Reprovado.

* Nota entre 5 e 6.9: Recuperação.

* Nota entre 7 e 8.9: Bom.

* Nota entre 9 e 10: Excelente.

* Nota fora do intervalo (menor que 0 ou maior que 10): Nota inválida.

```python
nota = float(input("Digite a nota do aluno (0 a 10): "))

if nota < 5: 
    print ("Reprovado.")
elif nota < 6.9:
    print ("Recuperação")
elif nota < 8.9:
    print ("Bom")
elif nota <= 10:
    print ("Excelente")
else:
    print("Nota Inválida") 
```

### E se quisermos repetir uma decisão várias vezes?

mas o que acontece quando precisamos verificar algo repetidamente? Por exemplo, se quisermos verificar um sinal de trânsito a cada segundo ou processar múltiplos elementos de uma lista?



É aí que entram os *loops*!

# Loops em Python

Os loops nos permitem repetir uma ação várias vezes sem precisar escrever o mesmo código repetidamente. Existem duas maneiras de criar loops em Python:

* _for_ → Utilizamos quando sabemos quantas vezes queremos repetir uma ação.
* _while_ → Executa repetidamente um bloco de código enquanto uma condição for verdadeira. Diferente do for, ele não tem um número fixo de repetições, mas continua até que a condição seja falsa. 

## _for_ loop:

Como dito anteriormente, o for é ideal quando sabemos quantas vezes queremos repetir uma ação. Ele percorre elementos de uma sequência, como listas, strings ou intervalos numéricos.

#### Exemplo
Imprima os números de 1 a 5 usando um loop _for_.

```python
for i in range(1,6):
    print (f"iteração {i}")
```

Se a gente fosse traduzir para a linguagem natural, teríamos: _"Para cada número de 1 a 5, faça o seguinte: imprima esse número."_

Agora vamos ver esse mesmo exemplo utilizando o for _while_?

## _while_ loop:

O while não tem um número fixo de repetições. Ele continua executando enquanto uma condição for verdadeira. Seria algo tipo esse diagrama de fluxo abaixo:  



#### Exemplo
Imprima os números de 1 a 5 usando um loop _while_.

```python
contador = 1

while contador <= 5:
    print (f"Iteração {contador}")
    contador = contador + 1
```

Sacou? Portanto, o loop while é útil quando *não sabemos exatamente quantas vezes o código deve rodar* e queremos que ele continue até que uma condição específica seja atendida.
