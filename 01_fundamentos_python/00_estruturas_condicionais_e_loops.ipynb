## Instru√ß√£o IF

A instru√ß√£o `if` √© uma das estruturas condicionais mais fundamentais na programa√ß√£o. Ela √© usada para tomar decis√µes com base em uma condi√ß√£o booleana (verdadeira ou falsa). Como vai funcionar:

* Se a condi√ß√£o for verdadeira (`true`), o bloco de c√≥digo que est√° _indentado_ abaixo da instru√ß√£o `if` ser√° executado.
* Caso a condi√ß√£o seja falsa (`false`), o bloco de c√≥digo dentro do `if` ser√° ignorado e o programa continuar√° a execu√ß√£o normalmente.


Ou seja, o if permite que um bloco de c√≥digo seja executado se uma condi√ß√£o for verdadeira.

```python
x=30

if x>15:
    print("x √© maior que 15")
```

## Instru√ß√£o ELIF

Se a condi√ß√£o do if for falsa, o python verifica a pr√≥xima condi√ß√£o do elif e assim por diante... sacou?
Vamos pensar num exemplo ent√£o?
Vamos elaborar um c√≥digo onde vamos classificar uma pessoa de acordo sua faixa et√°ria:
* Crian√ßa: At√© 12 anos.
* Adolescente: Entre 12 e 17 anos.

```python
idade = 22

if idade <12:
    print("Crian√ßa")
elif idade <18:
    print ("Adolescente")
elif idade <60:
    print("Adulto")
elif idade >=60:
    print("Idoso")
```

## A instru√ß√£o ELSE

A instru√ß√£o else serve para pegar todas as situa√ß√µes que n√£o foram atendidas pelas condi√ß√µes anteriores (os comandos if e elif).  

Ou seja, **ela √© como uma "√∫ltima chance" de executar algo, caso nenhuma das condi√ß√µes anteriores tenha sido verdadeira.**

Ela n√£o precisa de uma condi√ß√£o para ser verificada; o bloco de c√≥digo que est√° dentro do else vai rodar automaticamente se todas as condi√ß√µes anteriores derem errado.  

Podemos reescrever o exemplo anterior da seguinte forma utilizando o _else_:

```python
idade = 70

if idade < 12:
    print("Crian√ßa")
elif idade < 18:
    print("Adolescente")
elif idade < 30:
    print("Jovem")
elif idade < 60:
    print("Adulto")
else:
    print("Idoso")
```

Vamos pensar em mais um exemplo s√≥ pra fechar com chave de ouro?  

Vamos criar um c√≥digo em python utilizando _if_, _elif_ e _else_ para classificar o desempenho de um aluno com base na sua nota e considerando as faixas abaixo:  

* Nota abaixo de 5: Reprovado.

* Nota entre 5 e 6.9: Recupera√ß√£o.

* Nota entre 7 e 8.9: Bom.

* Nota entre 9 e 10: Excelente.

* Nota fora do intervalo (menor que 0 ou maior que 10): Nota inv√°lida.

```python
nota = float(input("Digite a nota do aluno (0 a 10): "))

if nota < 5: 
    print ("Reprovado.")
elif nota < 6.9:
    print ("Recupera√ß√£o")
elif nota < 8.9:
    print ("Bom")
elif nota <= 10:
    print ("Excelente")
else:
    print("Nota Inv√°lida") 
```

### E se quisermos repetir uma decis√£o v√°rias vezes?

mas o que acontece quando precisamos verificar algo repetidamente? Por exemplo, se quisermos verificar um sinal de tr√¢nsito a cada segundo ou processar m√∫ltiplos elementos de uma lista?



√â a√≠ que entram os *loops*!

Os loops nos permitem repetir uma a√ß√£o v√°rias vezes sem precisar escrever o mesmo c√≥digo repetidamente. Existem duas maneiras de criar loops em Python:

* _for_ ‚Üí Utilizamos quando sabemos quantas vezes queremos repetir uma a√ß√£o.
* _while_ ‚Üí Executa repetidamente um bloco de c√≥digo enquanto uma condi√ß√£o for verdadeira. Diferente do for, ele n√£o tem um n√∫mero fixo de repeti√ß√µes, mas continua at√© que a condi√ß√£o seja falsa. 

## _for_ loop:

Como dito anteriormente, o for √© ideal quando sabemos quantas vezes queremos repetir uma a√ß√£o. Ele percorre elementos de uma sequ√™ncia, como listas, strings ou intervalos num√©ricos.

#### Exemplo
Imprima os n√∫meros de 1 a 5 usando um loop _for_.

```python
for i in range(1,6):
    print (f"itera√ß√£o {i}")
```

Se a gente fosse traduzir para a linguagem natural, ter√≠amos: _"Para cada n√∫mero de 1 a 5, fa√ßa o seguinte: imprima esse n√∫mero."_

Agora vamos ver esse mesmo exemplo utilizando o for _while_?

## _while_ loop:

O while n√£o tem um n√∫mero fixo de repeti√ß√µes. Ele continua executando enquanto uma condi√ß√£o for verdadeira.

#### Exemplo
Imprima os n√∫meros de 1 a 5 usando um loop _while_.

```python
contador = 1

while contador <= 5:
    print (f"Itera√ß√£o {contador}")
    contador = contador + 1
```

Sacou? Portanto, o loop while √© √∫til quando *n√£o sabemos exatamente quantas vezes o c√≥digo deve rodar* e queremos que ele continue at√© que uma condi√ß√£o espec√≠fica seja atendida.
